```python
#Importando biblioteca numpy
import numpy as np
```

```python
#Carregando url que contem a base de dados
url = 'https://raw.githubusercontent.com/alura-cursos/numpy/dados/apples_ts.csv'
```

```python
#Carregando os dados através de uma URL, usando o comando delimiter para identificar o separador, e o usecols para informar quais colunas serão usadas
#np.arange cria um array que irá ser a quantidade de colunas que serão usadas, 1 é o indice que irá começar, 88 o numero total de colunas, o segundo 1 é o incremento
dado = np.loadtxt(url, delimiter=',', usecols=np.arange(1,88,1))
dado
```

```python
#Verificando e imprimindo o número de dimensões do array
#Retorna um dado bidimensional, ou seja, organizado por dois fatores, meses e cidades
dado.ndim
```

```python
#Verificando o numero de elementos no dado
dado.size
```

```python
#Verificando o formato do dados
#Dados com 6 linhas e 87 colunas
dado.shape
```

```python
#Realizando a transposição dos dados, transformando linhas em colunas e vice-versa
dados_transposto = dado.T
dados_transposto
```

##Operações entre arrays


```python
'''
y = 2 * x + 80
y é o preço das maçãs (Váriavel dependente).
x é o tempo (meses). (Váriavel independente)
2 (numero aleatório, apenas para exemplo) é o quanto o preço aumenta a cada mês (coeficiente angular).
80 é o preço inicial (ponto de partida, intercepto da reta).
'''
#Usarei o np.polyfit para calcular o valor do coeficiente angular, nesse exemplo, o quanto o preço aumenta a cada mês. Também para calcular o ponto de partida
a,b = np.polyfit(datas, Moscow,1)
y=a*datas+b

plt.plot(datas, Moscow)
plt.plot(datas, y)
```

```python
#Verificando o quanto a reta se ajusta aos dados
np.linalg.norm(Moscow-y)
```

###Fazendo o mesmo calculo, mas de forma manual

â = coeficiente angular;<br>
n = número de elementos;<br>
Y = Moscow;<br>
X = datas.<br>

```python
#Atribuindo os valores para as formulas
y_manual = Moscow
x_manual = datas
n_manual = np.size(Moscow)
```

```python
#Fazendo a conta do coeficiente angular - a
a_manual = (n_manual*np.sum(x_manual*y_manual) - np.sum(x_manual)*np.sum(y_manual))/(n_manual*np.sum(x_manual**2) - np.sum(x_manual)**2)
```

image.png




```python
#Fazendo a conta do coeficiente linear - b
b_manual = np.mean(y_manual) - a_manual*np.mean(x_manual)
b_manual
```

```python
y_manual = a_manual*x_manual+b_manual
```

```python
#Ao verificar o ajuste da reta aos dados, vemos que tem o mesmo valor que fomos retornados ao usar o numpy para calcular os valores
np.linalg.norm(Moscow-y_manual)
```
