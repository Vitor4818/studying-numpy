```python
#Importando biblioteca numpy
import numpy as np
```

```python
#Carregando url que contem a base de dados
url = 'https://raw.githubusercontent.com/alura-cursos/numpy/dados/apples_ts.csv'
```

```python
#Carregando os dados através de uma URL, usando o comando delimiter para identificar o separador, e o usecols para informar quais colunas serão usadas
#np.arange cria um array que irá ser a quantidade de colunas que serão usadas, 1 é o indice que irá começar, 88 o numero total de colunas, o segundo 1 é o incremento
dado = np.loadtxt(url, delimiter=',', usecols=np.arange(1,88,1))
dado
```

```python
#Verificando e imprimindo o número de dimensões do array
#Retorna um dado bidimensional, ou seja, organizado por dois fatores, meses e cidades
dado.ndim
```

```python
#Verificando o numero de elementos no dado
dado.size
```

```python
#Verificando o formato do dados
#Dados com 6 linhas e 87 colunas
dado.shape
```

```python
#Realizando a transposição dos dados, transformando linhas em colunas e vice-versa
dados_transposto = dados.T
dados_transposto
```

```python
#Separando por cidades
Moscow = precos[:,0]
Kaliningrad = precos[:,1]
Petersburg = precos[:,2]
Krasnodar = precos[:,3]
Ekaterinburg = precos[:,4]
```

##Comparação entre arrays

```python
Moscow.shape
```

```python
#vamos verificar se a epóca do ano impacta nos preços das maças
Moscow_ano1 = Moscow[0:12]
Moscow_ano2 = Moscow[12:24]
Moscow_ano3 = Moscow[24:36]
Moscow_ano4 = Moscow[36:48]
```

```python
#Criando um array que contem 12 elementos, ou seja, a quantidade de meses em um ano
ano = np.arange(1,13,1)
```

```python
#Visualizando os dados
plt.figure(figsize=(10,5))
plt.plot(ano, Moscow_ano1)
plt.plot(ano, Moscow_ano2)
plt.plot(ano, Moscow_ano3)
plt.plot(ano, Moscow_ano4)
plt.legend(['Primeiro ano', 'Segundo ano', 'Terceiro ano', 'Quarto ano'])
```

```python
#Comparando o array para ver se são iguais
np.array_equal(Moscow_ano3,Moscow_ano4)
```

```python
#Verificando a diferença entre os array
#Se a diferença entre os elementos for maior do que a tolerância, o método retorna False
#Caso contrário, retorna true
np.allclose(Moscow_ano3,Moscow_ano4, 0.1)
```
